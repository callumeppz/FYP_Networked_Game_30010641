# Game Sales Overtime #

```python
CSV = pd.read_csv(r'C:\Users\Callum Apps\Downloads\Gamesales.csv\GameSales.csv')
CSV.head()
```

```python
drop_row_index = CSV[CSV['Year'] > 2015].index
CSV = CSV.drop(drop_row_index)
```

# Growth in Global sales over the years

```python
data_year = CSV.groupby(by=['Year'])['Global_Sales'].sum().reset_index()

plt.figure(figsize=(12, 8))
sns.barplot(x="Year", y="Global_Sales",data=data_year,  palette='Blues')
plt.xticks(rotation=90)

plt.xlabel('Year Of Sales')
plt.ylabel('Global Sales')
plt.title('Global Sales Count by Year')

plt.show()
```

# Most Popular Genre

```python
# Group by Year and count the occurrences of each Genre
data_year = CSV.groupby(by=['Genre'])['Global_Sales'].sum().reset_index()

# Set the plotting style
sns.set(style="whitegrid")

# Create a bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x="Genre", y="Global_Sales", data=data_year, palette='Blues')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Show the plot
plt.show()
```
